```python
cc2
```

```python
ccall=pd.concat([cc1, cc2, cc3], sort=True, ignore_index=True)
```

```python
ccall #자동 인덱싱 붙여서 해
```

```python
ccall.iloc[1,1] 
```

```python
c1=ccall.loc[2] #2행 2번 row
```

```python
c2= ccall.iloc[:,0] #0열 0번 칼럼
```

```python
c4
```
